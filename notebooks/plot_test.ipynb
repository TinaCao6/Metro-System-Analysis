```python
OUT_PATH = "/tmp/output"
DAY = "monday"
AVG_WALKING_SPEED = 1.4 # 1.4 meters per second
MAX_WALKING_MIN = 12
geolocator = Nominatim(user_agent="wisc_bazarr_demo")
```

```python
# Charles
DATA_PATH = "mygtfs.zip"
START_TIME = "04:50:00"
ELAPSE_TIME = "01:00:00"
start_location = "330 N Orchard St"

# Tina
DATA_PATH_2 = "../data/transfer_gtfs.zip"
START_TIME_2 = "12:58:47"
ELAPSE_TIME_2 = "00:59:00"
start_location_2 = "The Lux"

# Young
DATA_PATH_3 = "../data/loop_gtfs_young.zip"
START_TIME_3 = "10:00:00"
ELAPSE_TIME_3 = "02:00:00"

# Celia
DATA_PATH_4 = "../data/loop_gtfs.zip"
START_TIME_4 = "10:00:00"
ELAPSE_TIME_4 = "02:00:00"
start_location_4 = "sellery"
```

```python
def drawn_bubble(start_point=None, start_location=None, busSim=None, crs=3174):
    location = geocode(start_location)
    lat, lon = (location.latitude, location.longitude)
    gdf = busSim.get_gdf(start_point=(lat, lon))
    gdf = gdf.to_crs(epsg=3174)
    bubble = flatten(gdf.geometry)
    bubble.boundary.plot(color='orange')
```

```python
# use argument to detect validity of different city
# Charles
print("check Charles' gtfs")
busSim1 = gen_busSim(DATA_PATH, OUT_PATH, DAY, START_TIME, ELAPSE_TIME, AVG_WALKING_SPEED, MAX_WALKING_MIN)
area1 = get_area(start_location=start_location, busSim=busSim1)
stops_radius_list = get_stops_radius_list(busSim=busSim1)
print('*' * 50)
drawn_bubble(start_location=start_location, busSim=gen_busSim(DATA_PATH, OUT_PATH, DAY, START_TIME, ELAPSE_TIME, AVG_WALKING_SPEED, MAX_WALKING_MIN))
get_error(DATA_PATH)
print(f'{area1=}')
print(f'{stops_radius_list}')
```
