```python
import json
import pandas as pd
import geopandas as gp
import matplotlib
import gtfs_kit as kit
```

```python
feed = kit.feed.read_feed('../data/mmt_gtfs.zip', dist_units='km')
```

```python
kit.validators.validate(feed, as_df=True, include_warnings=True)
```

```python
feed = feed.drop_invalid_columns()
feed = feed.clean()
```

```python
first_week = feed.get_first_week()
first_week
```
