```python
import json
import pandas as pd
import geopandas as gp
import matplotlib
import gtfs_kit as kit
import matplotlib.pyplot as plt
```

```python
feed = kit.feed.read_feed('../data/mmt_gtfs.zip', dist_units='km')
kit.validators.validate(feed, as_df=True, include_warnings=False)
feed = feed.drop_invalid_columns().clean()
week = feed.get_week(1)
feed.describe()
```

```python
feed.describe().to_markdown()
```

```python
routes = feed.get_routes(date=week[0])
trip_stats = feed.compute_trip_stats()
fts = feed.compute_feed_time_series(trip_stats, dates=week, freq="2H")
feed_stats = feed.compute_feed_stats(trip_stats, week)
rts = feed.compute_route_time_series(trip_stats, week, freq='12H')
```

```python
inds = ['service_distance', 'service_duration']
rids = ['9500', '9502']

simple_rts = rts.loc[:, (inds, )]
```

```python
simple_rts
```

```python
min_sdis = simple_rts['service_distance'].describe().loc['min']
```

```python
min_stime = simple_rts['service_duration'].describe().loc['min']
```

```python
min_sdis.sort_values()[:5].plot.bar(xlabel="Route Id", ylabel="Service Distance (km)")
plt.savefig("madison_min_service_dist.png",dpi=200, bbox_inches='tight')
```

```python
min_stime.sort_values()[:5].plot.bar(xlabel="Route Id", ylabel="Service duration(h)")
plt.savefig("madison_min_service_time.png",dpi=200, bbox_inches='tight')
```
